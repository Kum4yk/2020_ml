```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# conda install ffmpeg
```

```python
test_file_path = "../data/motorama-wind_in_her_hair.mp3"
test_filename = "motorama-wind_in_her_hair.mp3"
```

```python
y, sr = librosa.load(test_file_path)
D = librosa.stft(y)
```

```python
y_db_max = librosa.amplitude_to_db(D, ref=np.max)
y_db = librosa.amplitude_to_db(D)
```

```python
def plot_figure(y, file_name, dpi=150):
    plt.figure(dpi=dpi)
    
    librosa.display.specshow(
        y,
        y_axis='log',
        x_axis='time'
    )

    plt.title(f'Power spectrogram of {file_name}')
    plt.colorbar(format='%+2.0f dB')
    plt.tight_layout()
```

```python
%%time
plot_figure(y_db, test_filename)
```

```python
plot_figure(y_db_max, test_filename)
```

```python
plot_figure(D, test_filename)
```
